```python
#we'd like to trim down some of the outliers based on price
from scipy import stats
import numpy as np
def remove_outliers_by_category(df, category_column, value_column, z_threshold):
    grouped = df.groupby(category_column)
    df_outliers_removed = pd.DataFrame(columns = df.columns)
    
    for category, group in grouped:
        mask = np.abs(stats.zscore(group[value_column])) < z_threshold
        df_outliers_removed = pd.concat([df_outliers_removed, group[mask]])
        
    return df_outliers_removed
```

```python
housing_data_outliers_removed = remove_outliers_by_category(housing_data, 'state', 'price', 3)
```
